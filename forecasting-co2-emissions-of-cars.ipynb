```python
X=df[["Engine Size(L)","Cylinders","Fuel Consumption City (L/100 km)","Fuel Consumption Hwy (L/100 km)","Fuel Consumption Comb (L/100 km)","Fuel Consumption Comb (mpg)"]]

y=df[["CO2 Emissions(g/km)"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.2, train_size=0.8,random_state=42)
```

```python
y.shape
```

```python
from xgboost.sklearn import XGBRegressor
from sklearn.linear_model import LinearRegression
```

```python
linear_regression= LinearRegression()
xgb_model=XGBRegressor()
```

```python
linear_regression.fit(X_train,y_train)
xgb_model.fit(X_train,y_train)
```

```python
linear_score = linear_regression.score(X, y)
xgb_score=xgb_model.score(X,y)
print('score of linear square: ', linear_score)
print('score of xgb square: ', linear_score)


#coeff_df = pd.DataFrame(linear_regression.coef_, X.columns,columns=['Coefficient'])
#coeff_df
#print('intercept:', linear_regression.intercept_)
```

```python
print(linear_regression.intercept_, linear_regression.coef_[0])
```

```python
from sklearn.metrics import r2_score
y_pred = linear_regression.predict(X_test)
r2 = r2_score(y_test, y_pred)
r2
```
