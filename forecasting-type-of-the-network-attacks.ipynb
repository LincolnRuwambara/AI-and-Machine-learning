```python
X=df.drop('Category', axis=1, inplace=False)
y=df['Category']


```

**Split your dataset as %80 training and %20 testing**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

train_size = 0.80
test_size = 0.20
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 666, test_size=test_size)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_scale, y, random_state = 666, test_size=test_size)


print(X_train.shape)
print(X_test.shape)
```

**Implement four classification model based on Logistic Regression, Support Vector Machine,
Multinomial Naive Bayes and Random Forest classifiers.**

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier

```

**Train (fit) your network.******

```python
models = [LogisticRegression(multi_class = 'auto', solver = 'liblinear', max_iter = 10000, C = 10), 
          svm.SVC(), 
          MultinomialNB(), 
          RandomForestClassifier(n_estimators = 200, max_depth = 30)]

```

****Report the accuracies (by percentage) of the models for the test datasets.****

```python
scores = {}
for i in range(len(models)):
    model = models[i]
    if i == 2:              
      model.fit(X_train, y_train)
      y_pred = model.predict(X_test)
      accuracy = accuracy_score(y_test, y_pred)
    else:                  
      model.fit(X_train2, y_train2)
      y_pred = model.predict(X_test2)
      accuracy = accuracy_score(y_test2, y_pred)
    
    scores[i] = accuracy * 100
    print(f'Accuracy of the model {models[i]}: {accuracy * 100}')
```
